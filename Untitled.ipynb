```python
rolling_avg = browser_trends.rolling(6).mean()
```

```python
rolling_avg['Month'] = browser_trends['Month']
```

```python
rolling_avg = rolling_avg[['Month', 'Firefox', 'Safari', 'Google Chrome', 'Internet Explorer', 'Opera']]
```

```python
rolling_avg.set_index('Month', inplace= True)
```

```python
rolling_avg.plot(figsize= (10, 8), colormap= 'gist_rainbow', grid= True)

```

```python
copy = browser_trends.copy()
copy.set_index('Month', inplace= True)
```

```python
pct_change_quarterly_copy = copy.pct_change(periods= 4)
```

```python
pct_change_quarterly_copy = pct_change_quarterly_copy[pct_change_quarterly_copy.index.year >= 2009]*100
```

```python
pct_change_quarterly_copy.head()
```

```python
pct_change_quarterly_copy.plot(subplots= True, title = ['Percentage change quarterly using google trend data Firefox', 
                                                        'Percentage change quarterly using google trend data Safari',
                                                        'Percentage change quarterly using google trend data Chrome',
                                                        'Percentage change quarterly using google trend data Internet Explorer',
                                                        'Percentage change quarterly using google trend data Opera'],
                               fontsize= 14, figsize= (20, 20), grid= True, yticks= [-30, -20, -10, 0, 10, 20, 30, 40, 50])
#plt.savefig('Google trends on browsers pct change quarterly.jpg', dpi= 80, quality= 90)
```
